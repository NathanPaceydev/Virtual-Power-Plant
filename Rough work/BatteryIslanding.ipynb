## Battery Islanding Calcs

```python
DesiredConsumption = 200 # hourly demand kW usage
runtime = 10 # number of hours customer wants system to be able to island for
percentBat = 50
batLife = [25,23,21,19,17,15,13]

# Note: the average outage in Ontario urban areas is 2 hours

print('Tesla Megapack provides 3.9 MWh of energy storage \n')
batCap = round(DesiredConsumption * runtime,2) # calculate the min battery capacity [kWh]

if percentBat == 100:
    
    batCapMWh = round(batCap/1000,2) # calculate the min battery capacity [MWh]
    # print capacities
    print('Smallest battery capacity for use case is:')
    print(batCap,' [kWh]')
    print(batCapMWh,' [MWh]\n')
    print('Approximate battery life:')
    print(batLife[0],' [yrs]')

    
elif percentBat == 90:
    # 90% battery energy usage
    cap90 = round(batCap/0.9,2) # calculate total battery capacity [kWh]
    cap90MWh = round(cap90/1000,2) # calculate total battery capacity [MWh]
    # print capacities
    print('Total battery capacity for 90% of overall battery usage')
    print(cap90,' [kWh]')
    print(cap90MWh,' [MWh]\n')
    print('Approximate battery life:')
    print(batLife[1],' [yrs]')

    
elif percentBat == 80:
    # 80% battery energy usage
    cap80 = round(batCap/0.8,2) # calculate total battery capacity [kWh]
    cap80MWh = round(cap80/1000,2) # calculate total battery capacity [MWh]
    # print capacities
    print('Total battery capacity for 80% of overall battery usage')
    print(cap80,' [kWh]')
    print(cap80MWh,' [MWh]\n')
    print('Approximate battery life:')
    print(batLife[2],' [yrs]')


elif percentBat == 70:
    # 70% battery energy usage
    cap70 = round(batCap/0.7,2) # calculate total battery capacity [kWh]
    cap70MWh = round(cap70/1000,2) # calculate total battery capacity [MWh]
    # print capacities
    print('Total battery capacity for 70% of overall battery usage')
    print(cap70,' [kWh]')
    print(cap70MWh,' [MWh]\n')
    print('Approximate battery life:')
    print(batLife[3],' [yrs]')


elif percentBat == 60:
    # 60% battery energy usage
    cap60 = round(batCap/0.6,2) # calculate total battery capacity [kWh]
    cap60MWh = round(cap60/1000,2) # calculate total battery capacity [MWh]
    # print capacities
    print('Total battery capacity for 60% of overall battery usage')
    print(cap60,' [kWh]')
    print(cap60MWh,' [MWh]\n')
    print('Approximate battery life:')
    print(batLife[4],' [yrs]')


elif percentBat == 50:
    # 50% battery energy usage
    cap50 = round(batCap/0.5,2) # calculate total battery capacity [kWh]
    cap50MWh = round(cap50/1000,2) # calculate total battery capacity [MWh]
    # print capacities
    print('Total battery capacity for 50% of overall battery usage')
    print(cap50,' [kWh]')
    print(cap50MWh,' [MWh]\n')
    print('Approximate battery life:')
    print(batLife[5],' [yrs]')


elif percentBat == 40:
    # 40% battery energy usage
    cap40 = round(batCap/0.4,2) # calculate total battery capacity [kWh]
    cap40MWh = round(cap40/1000,2) # calculate total battery capacity [MWh]
    # print capacities
    print('Total battery capacity for 40% of overall battery usage')
    print(cap40,' [kWh]')
    print(cap40MWh,' [MWh]')
    print('Approximate battery life:')
    print(batLife[6],' [yrs]')
```
